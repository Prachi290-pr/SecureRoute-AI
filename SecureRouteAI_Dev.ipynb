# SecureRoute-AI Dev Notebook

Quick notebook for validating environment behavior and running baseline inference.

```python
import os
from environment import SecureRouteEnv
from models import Action, RoutingDepartment
from graders import grade_easy, grade_medium, grade_hard
```

```python
env = SecureRouteEnv()
obs = env.reset(ticket_id=3)
print(f'Ticket ID: {obs.ticket_id}')
print(obs.text)
```

```python
# Example medium-task action for ticket 3
redacted = obs.text.replace('4111-1111-1111-1111', '[REDACTED]')
action = Action(redacted_text=redacted, routing=RoutingDepartment.BILLING)
_, reward, done, info = env.step(action)
print('Done:', done)
print('Score:', reward.score)
print('Reason:', reward.reason)
```
